```python
import pandas as pd
from Bio.SeqIO import parse
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq
```

```python
csv_contents = pd.read_csv("sequences.csv")
print(csv_contents.iloc[1,0])
fasta_file = open("sequences.fasta")
records = parse(fasta_file,"fasta")
```

```python
seq_list=[]
count=0
for record in records:
    if csv_contents.iloc[count,0] == str(record.id):
        print(f'{record.id} found in the Fasta.')
        seq_list.append(str(record.seq))
    count += 1
print(count)
```
